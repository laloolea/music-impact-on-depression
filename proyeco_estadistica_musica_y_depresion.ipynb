# **Introducción** 


<t align="justify">

En los últimos años, ha habido un creciente interés en comprender la relación entre la música y la salud mental. Varios estudios han demostrado los efectos positivos de la música en el bienestar psicológico y la reducción de los síntomas de diversas condiciones de salud mental (Dingle et al., 2021; Román, 2022). Además, se ha observado que la música puede desempeñar un papel importante en el manejo del estrés y la promoción del bienestar general.




### Antecedentes 



<t align="justify">

Investigaciones anteriores han examinado los efectos terapéuticos de la música en diferentes contextos. Por ejemplo, Dingle et al. (2021) realizaron una revisión exhaustiva de estudios que exploraron los mecanismos psicosociales mediante los cuales las actividades musicales afectan la salud y el bienestar. Los resultados de esta revisión destacaron la capacidad de la música para mejorar el estado de ánimo, reducir el estrés y promover la expresión emocional en diversas poblaciones.

Además, Román (2022) investigó los efectos de la música en el bienestar y la salud mental. El autor encontró evidencia de que la música puede tener un impacto positivo en la regulación emocional, la reducción de depresión y el fomento de la relajación. Estos hallazgos respaldan la idea de que la música puede ser una herramienta eficaz en el cuidado de la salud mental.

Sin embargo, a pesar de los avances en la comprensión de los efectos terapéuticos de la música, existe una brecha de conocimiento en cuanto a su relación específica con la depresión. Se requiere una investigación más profunda para comprender mejor cómo la música puede influir en la incidencia y la gravedad de la depresión.

### **Planteamiento del Problema**

<t align="justify">

La depresión es un trastorno mental grave que afecta a un gran número de personas en todo el mundo. Se ha demostrado que la música puede tener un impacto significativo en el bienestar mental y emocional de las personas. Existen estudios que sugieren que la música puede influir en el estado de ánimo, reducir el estrés y mejorar la calidad de vida de las personas que experimentan síntomas de depresión (Dingle et al., 2021; Román, 2022). Sin embargo, aún hay aspectos que no se han explorado a fondo.

### **Pruebas Estadísticas**

#### Correlación entre horas al día escuchando música y grado de depresión reportado

<t align="justify">

Como primer paso, independientemente del género musical, nos gustaría analizar si existe una correlación entre la cantidad de música escuchada al día y el grado de depresión reportado. Para esto, realizamos una prueba de correlación entre esta dos variables y efectuamos una prueba de significancia. 

```python
from scipy.stats import pearsonr

# Seleccionar las columnas de interés
hours_per_day = df['Hours per day']
Depression = df['Depression']

# Eliminar filas con valores faltantes en alguna de las columnas
df_clean = df[['Hours per day', 'Depression']].dropna()

# Calcular la correlación
correlation, p_value = pearsonr(df_clean['Hours per day'], df_clean['Depression'])

# Imprimir el coeficiente de correlación y el p-valor
print("Coeficiente de correlación:", correlation)
print("P-valor:", p_value)

# Generar el histograma 2D con cmap='hot'
plt.hist2d(df_clean['Hours per day'], df_clean['Depression'], bins=[20, 20], cmap='hot')

# Configurar los límites de los ejes x e y
plt.xlim(df_clean['Hours per day'].min(), df_clean['Hours per day'].max())
plt.ylim(df_clean['Depression'].min(), df_clean['Depression'].max())

# Configurar las etiquetas de los ejes x e y y el título del gráfico
plt.xlabel('Horas al día escuchando música')
plt.ylabel('Grado de depresión')
plt.title('Correlación entre horas al día escuchando música y grado de depresión')
plt.text(30,2.5,"Cantidad de personas",rotation="vertical")

# Añadir una barra de color (colorbar)
plt.colorbar()

# Mostrar el gráfico
plt.show()
```

<t align="justify">

De antemano sabiamos que mut pocos individuos en nuestra muestra dedican más de 5 horas al día a esuchcar música. Por lo tanto, es de esperarse que los datos se concentren en la parte izquierda de la figura. Concentrándonos en esta parte, notamos que el patrón observado no es informativo. Esto es porque, independientemente de la cantidad de horas al día que se dedica a escuchar música, dentro del rango de 0 a 5 horas, encontramos grados de depresión que van desde lo más bajo a lo más alto. En otras palabras, los grados de depresión reportados por estos individuos, no parece estar directamente correlacionados con la cantidad de música que se escucha al día. 

No obstante, nuestra prueba de correlación obtuvo un valor de p significativo (p < 0.05). Sin embargo, el hecho de que el coefienciente de correlación sea tan pequeño, vuelve díficil observar el supuesto efecto. Una muestra más grande podría, por un lado, volver más notorio dicho efecto, o por el otro, hacer que esta prueba no alcance la significancia. 

#### ANOVA entre el género músical favorito y el grado de depresión reportado

<t align="justify">

Habiendo corroborado que la música por sí misma no se correlaciona con la incidencia de un grado de depresión mayor o menor, procedemos a analizar si géneros músicales específicos sí están teniendo un efecto sobre los escuchas en cuanto a síntomas depresivos. Para esto, decidimos seleccionar solo a aquellos individuos que reportaron escuchar música una o más horas al día. 

```python
import matplotlib.style as style
from scipy.stats import f_oneway

# Establecer el estilo a 'ggplot'
style.use('ggplot')

# Filtrar los registros de las personas que escuchan música 1 hora al día o más
df_filtered = df[df['Hours per day'] >= 1]

# Seleccionar las columnas de interés
fav_genre = df_filtered['Fav genre']
depression = df_filtered['Depression']

# Eliminar filas con valores faltantes en alguna de las columnas
df_clean = df_filtered[['Fav genre', 'Depression']].dropna()

# Calcular el grado de depresión promedio por género favorito
avg_depression = df_clean.groupby('Fav genre')['Depression'].mean().reset_index()

# Generar el gráfico de caja y bigotes
plt.figure(figsize=(16, 6))
plt.boxplot([group_data for _, group_data in df_clean.groupby('Fav genre')['Depression']])

# Configurar las etiquetas de los ejes x e y y el título del gráfico
plt.xlabel('Género Favorito')
plt.ylabel('Grado de Depresión')
plt.title('Relación entre género favorito y depresión')

# Establecer las etiquetas del eje x
plt.xticks(range(1, len(avg_depression) + 1), avg_depression['Fav genre'], rotation=45)

# Realizar el análisis de ANOVA
groups = []
for group_name, group_data in df_clean.groupby('Fav genre')['Depression']:
    groups.append(group_data.tolist())

f_statistic, p_value = f_oneway(*groups)

# Imprimir los resultados del análisis de ANOVA
print("Estadístico F:", f_statistic)
print("P-valor:", p_value)

# Mostrar el gráfico
plt.show()
```

<t align="justify">

La prueba de ANOVA no es significativa, lo cual se denota en el hecho de que todas las cajas presentan un cierto grado de solapamiento. No obstante, el valor de p de la prueba fue de 0.06, muy cercano a la significancia. Una posible causa de esto puede ser el género Lofi. Al observar su gráfica, notamos que su mediana se encuentra casi por encima del tercer cuartil del resto de los géneros. Esto implica que la mitad de los individuos que seleccionaron Lofi como genero favorito, poseen un grado de depresión que esta por encima de las tres cuartas partes de los grados de depresión reportados en el resto de los géneros.  

Nota: debemos tener en cuenta que Lofi es uno de los generos que fue seleccionado como favorito en muy pocas ocasiones, por lo que cualquier efecto observado en estos datos debe de ser interpretado con precaución. 

<t align="justify">

A continuación, para obtener más información sobre la influencia de escuchar géneros particulares sobre el grado de depresión reportado decidimos utilizar un abordaje un poco distinto. Habiendo filtrado a aquellos individuos que escuchan por lo menor una hora de música al día, para cada género, seleccionamos a aquellos individuos que reportaron escucharlo "Sometimes" o "Very frequently". Esto nos permite analizar el grado de depresión de aquellas personas que, independientemente de su género favorito, escuchan ciertos géneros con cierta frecuencia. Realizamos nuevamente una ANOVA y prueba de significancia en estos grupos. 

```python
# Filtrar los registros de las personas que escuchan música 1 hora al día o más
df_filtered = df[df['Hours per day'] >= 1]

# Crear una lista para almacenar los datos de depresión de cada género
depression_data = []

# Realizar el filtrado secuencial para cada género
for column in df_filtered.columns[12:27]:
    # Filtrar los sujetos que respondieron "sometimes" o "very frequently" para el género actual
    filtered_data = df_filtered[df_filtered[column].isin(['Sometimes', 'Very frequently'])]

    # Obtener los valores de depresión para el género actual
    depression_values = filtered_data['Depression'].values
    depression_data.append(depression_values)

# Crear un DataFrame con los géneros y los datos de depresión
df_depression = pd.DataFrame(depression_data).T
df_depression.columns = [column.split("[")[1][:-1] for column in df_filtered.columns[12:27]]

# Verificar si el DataFrame está vacío
if df_depression.empty:
    print("No hay suficientes datos para generar la gráfica.")
else:
    # Graficar los resultados utilizando un gráfico de caja y bigotes
    plt.figure(figsize=(12, 6))
    sns.boxplot(data=df_depression)

    # Configurar las etiquetas del eje x y el título del gráfico
    plt.xticks(rotation=45)
    plt.xlabel('Género Musical')
    plt.ylabel('Grado de Depresión')
    plt.title('Relación entre géneros musicales escuchados Sometimes o Very frequently y el grado de depresión')

    # Calcular el análisis de ANOVA
    f_statistic, p_value = f_oneway(*depression_data)

    # Imprimir los resultados del análisis de ANOVA
    print("Estadístico F:", f_statistic)
    print("P-valor:", p_value)

    # Mostrar el gráfico
    plt.show()
```

<t align="justify">

Aunque teniamos la esperanza de resaltar alguna de las ligeros efectos observados en el análisis anterior, nos topamos con el caso contrario: los grupos presentan distribuciones más similares entre sí. 
 

#### Análisis del grado de depresión en individuos cuyo género favorito es Lofi
