#### Correlacin entre horas al da escuchando msica y grado de depresin reportado

<t align="justify">

Como primer paso, independientemente del gnero musical, nos gustara analizar si existe una correlacin entre la cantidad de msica escuchada al da y el grado de depresin reportado. Para esto, realizamos una prueba de correlacin entre esta dos variables y efectuamos una prueba de significancia. 

```python
from scipy.stats import pearsonr

# Seleccionar las columnas de inters
hours_per_day = df['Hours per day']
Depression = df['Depression']

# Eliminar filas con valores faltantes en alguna de las columnas
df_clean = df[['Hours per day', 'Depression']].dropna()

# Calcular la correlacin
correlation, p_value = pearsonr(df_clean['Hours per day'], df_clean['Depression'])

# Imprimir el coeficiente de correlacin y el p-valor
print("Coeficiente de correlacin:", correlation)
print("P-valor:", p_value)

# Generar el histograma 2D con cmap='hot'
plt.hist2d(df_clean['Hours per day'], df_clean['Depression'], bins=[20, 20], cmap='hot')

# Configurar los lmites de los ejes x e y
plt.xlim(df_clean['Hours per day'].min(), df_clean['Hours per day'].max())
plt.ylim(df_clean['Depression'].min(), df_clean['Depression'].max())

# Configurar las etiquetas de los ejes x e y y el ttulo del grfico
plt.xlabel('Horas al da escuchando msica')
plt.ylabel('Grado de depresin')
plt.title('Correlacin entre horas al da escuchando msica y grado de depresin')
plt.text(30,2.5,"Cantidad de personas",rotation="vertical")

# Aadir una barra de color (colorbar)
plt.colorbar()

# Mostrar el grfico
plt.show()
```

<t align="justify">

De antemano sabiamos que mut pocos individuos en nuestra muestra dedican ms de 5 horas al da a esuchcar msica. Por lo tanto, es de esperarse que los datos se concentren en la parte izquierda de la figura. Concentrndonos en esta parte, notamos que el patrn observado no es informativo. Esto es porque, independientemente de la cantidad de horas al da que se dedica a escuchar msica, dentro del rango de 0 a 5 horas, encontramos grados de depresin que van desde lo ms bajo a lo ms alto. En otras palabras, los grados de depresin reportados por estos individuos, no parece estar directamente correlacionados con la cantidad de msica que se escucha al da. 

No obstante, nuestra prueba de correlacin obtuvo un valor de p significativo (p < 0.05). Sin embargo, el hecho de que el coefienciente de correlacin sea tan pequeo, vuelve dficil observar el supuesto efecto. Una muestra ms grande podra, por un lado, volver ms notorio dicho efecto, o por el otro, hacer que esta prueba no alcance la significancia. 

#### ANOVA entre el gnero msical favorito y el grado de depresin reportado

<t align="justify">

Habiendo corroborado que la msica por s misma no se correlaciona con la incidencia de un grado de depresin mayor o menor, procedemos a analizar si gneros msicales especficos s estn teniendo un efecto sobre los escuchas en cuanto a sntomas depresivos. Para esto, decidimos seleccionar solo a aquellos individuos que reportaron escuchar msica una o ms horas al da. 

```python
import matplotlib.style as style
from scipy.stats import f_oneway

# Establecer el estilo a 'ggplot'
style.use('ggplot')

# Filtrar los registros de las personas que escuchan msica 1 hora al da o ms
df_filtered = df[df['Hours per day'] >= 1]

# Seleccionar las columnas de inters
fav_genre = df_filtered['Fav genre']
depression = df_filtered['Depression']

# Eliminar filas con valores faltantes en alguna de las columnas
df_clean = df_filtered[['Fav genre', 'Depression']].dropna()

# Calcular el grado de depresin promedio por gnero favorito
avg_depression = df_clean.groupby('Fav genre')['Depression'].mean().reset_index()

# Generar el grfico de caja y bigotes
plt.figure(figsize=(16, 6))
plt.boxplot([group_data for _, group_data in df_clean.groupby('Fav genre')['Depression']])

# Configurar las etiquetas de los ejes x e y y el ttulo del grfico
plt.xlabel('Gnero Favorito')
plt.ylabel('Grado de Depresin')
plt.title('Relacin entre gnero favorito y depresin')

# Establecer las etiquetas del eje x
plt.xticks(range(1, len(avg_depression) + 1), avg_depression['Fav genre'], rotation=45)

# Realizar el anlisis de ANOVA
groups = []
for group_name, group_data in df_clean.groupby('Fav genre')['Depression']:
    groups.append(group_data.tolist())

f_statistic, p_value = f_oneway(*groups)

# Imprimir los resultados del anlisis de ANOVA
print("Estadstico F:", f_statistic)
print("P-valor:", p_value)

# Mostrar el grfico
plt.show()
```

<t align="justify">

La prueba de ANOVA no es significativa, lo cual se denota en el hecho de que todas las cajas presentan un cierto grado de solapamiento. No obstante, el valor de p de la prueba fue de 0.06, muy cercano a la significancia. Una posible causa de esto puede ser el gnero Lofi. Al observar su grfica, notamos que su mediana se encuentra casi por encima del tercer cuartil del resto de los gneros. Esto implica que la mitad de los individuos que seleccionaron Lofi como genero favorito, poseen un grado de depresin que esta por encima de las tres cuartas partes de los grados de depresin reportados en el resto de los gneros.  

Nota: debemos tener en cuenta que Lofi es uno de los generos que fue seleccionado como favorito en muy pocas ocasiones, por lo que cualquier efecto observado en estos datos debe de ser interpretado con precaucin. 

<t align="justify">

A continuacin, para obtener ms informacin sobre la influencia de escuchar gneros particulares sobre el grado de depresin reportado decidimos utilizar un abordaje un poco distinto. Habiendo filtrado a aquellos individuos que escuchan por lo menor una hora de msica al da, para cada gnero, seleccionamos a aquellos individuos que reportaron escucharlo "Sometimes" o "Very frequently". Esto nos permite analizar el grado de depresin de aquellas personas que, independientemente de su gnero favorito, escuchan ciertos gneros con cierta frecuencia. Realizamos nuevamente una ANOVA y prueba de significancia en estos grupos. 

```python
# Filtrar los registros de las personas que escuchan msica 1 hora al da o ms
df_filtered = df[df['Hours per day'] >= 1]

# Crear una lista para almacenar los datos de depresin de cada gnero
depression_data = []

# Realizar el filtrado secuencial para cada gnero
for column in df_filtered.columns[12:27]:
    # Filtrar los sujetos que respondieron "sometimes" o "very frequently" para el gnero actual
    filtered_data = df_filtered[df_filtered[column].isin(['Sometimes', 'Very frequently'])]

    # Obtener los valores de depresin para el gnero actual
    depression_values = filtered_data['Depression'].values
    depression_data.append(depression_values)

# Crear un DataFrame con los gneros y los datos de depresin
df_depression = pd.DataFrame(depression_data).T
df_depression.columns = [column.split("[")[1][:-1] for column in df_filtered.columns[12:27]]

# Verificar si el DataFrame est vaco
if df_depression.empty:
    print("No hay suficientes datos para generar la grfica.")
else:
    # Graficar los resultados utilizando un grfico de caja y bigotes
    plt.figure(figsize=(12, 6))
    sns.boxplot(data=df_depression)

    # Configurar las etiquetas del eje x y el ttulo del grfico
    plt.xticks(rotation=45)
    plt.xlabel('Gnero Musical')
    plt.ylabel('Grado de Depresin')
    plt.title('Relacin entre gneros musicales escuchados Sometimes o Very frequently y el grado de depresin')

    # Calcular el anlisis de ANOVA
    f_statistic, p_value = f_oneway(*depression_data)

    # Imprimir los resultados del anlisis de ANOVA
    print("Estadstico F:", f_statistic)
    print("P-valor:", p_value)

    # Mostrar el grfico
    plt.show()
```

<t align="justify">

Aunque teniamos la esperanza de resaltar alguna de las ligeros efectos observados en el anlisis anterior, nos topamos con el caso contrario: los grupos presentan distribuciones ms similares entre s.
